```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
# store feature and target in variables
feature = digits.data
target = digits.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
df=pd.DataFrame(digits['data'],columns=digits['feature_names'])
```

```python
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, target, train_size = 0.7, random_state = 3)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Apply Algorithm

```python
from sklearn.linear_model import LogisticRegression
```

```python
my_model = LogisticRegression()
```

```python
my_model.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
preds = my_model.predict(X_test)
print(accuracy_score(y_test, preds))
```

# Apply PCA

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=25)
x_pca=pca.fit_transform(df)
```

```python
df.shape
```

```python
explained_variance = np.var(x_pca, axis=0)
print(explained_variance)
```

```python
explained_variance_ratio = explained_variance / np.sum(explained_variance)
```

```python
PC_values = np.arange(pca.n_components) + 1
plt.plot(PC_values, explained_variance_ratio, 'o-', linewidth=2, color='blue')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.show()
```

```python
# Lists to store the results
accuracy_results = []
n_components_list = list(range(1, 60))
```

```python
# Lists to store the results
for n in n_components_list:
    pca = PCA(n_components=n)
    x_pca = pca.fit_transform(df)
    
    X_train, X_test, y_train, y_test = train_test_split(x_pca, target, train_size=0.7, random_state=3)
    
    my_model.fit(X_train, y_train)
    my_model_preds = my_model.predict(X_test)
    
    accuracy_results.append(accuracy_score(y_test, my_model_preds))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Plotting the results in a 1-column, 4-row layout
plt.figure(figsize=(10, 16))  # Adjust the figure size for better display

# MAE plot
plt.plot(n_components_list, accuracy_results, marker='o')
plt.title('accuracy_results vs n_components')
plt.xlabel('n_components')
plt.ylabel('accuracy_results')

# Adjust layout to avoid overlap
plt.tight_layout()
plt.show()
```

```python
best_n_accuracy = n_components_list[np.argmax(accuracy_results)]
best_accuracy = max(accuracy_results)

# Printing the result
print(f"Best n_components for Accuracy: {best_n_accuracy} with Accuracy = {best_accuracy}")
```
